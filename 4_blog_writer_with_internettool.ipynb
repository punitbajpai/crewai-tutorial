```python
from dotenv import load_dotenv
load_dotenv()
import os
os.environ["TAVILY_API_KEY"] = "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"
os.environ["LITELLM_LOG"] = "CRITICAL"
os.environ["LITELLM_DISABLE_LOGGING"] = "true"
os.environ["LITELLM_LOG_LEVEL"] = "CRITICAL"

from crewai import LLM

llm  = LLM(model="groq/llama-3.1-8b-instant",temperature=0, provider="groq")
```

```python
from crewai import Agent, Task, Crew
#from crewai_tools import Tool
from langchain_community.tools.tavily_search import TavilySearchResults

os.environ["TAVILY_API_KEY"] = "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"

search_tool = TavilySearchResults(max_results=3)


print(type(search_tool))

research_agent = Agent(
    role = "Research Specialist",
    goal = "Research interesting facts about the topic : {topic}",
    backstory = "You are an expert at finding relevant and factual data.",
    tools = [search_tool],
    verbose = True,
    llm = llm
)

writer_agent = Agent(
    role = "Creative Writer",
    goal = "Write a short blog summary using the research",1
    backstory = "You are skilled at writing engaging summaries based on provided content.",
    verbose = True,
    llm = llm
)

task1 = Task(
    description = "Find 3-5 interesting and recent facts about {topic} as of year 2025",
    expected_output="A bullest list of 3-5 facts",
    agent = research_agent,

)

task2 = Task(
    description = "Write a 100-word blog post summary about {topic} using the facts from the research.",
    expected_output="A blog post summary",
    agent = writer_agent,
    context = [task1],
)

crew = Crew(
    agents = [research_agent,writer_agent],
    tasks = [task1,task2],
    verbose = True,
)

crew.kickoff(inputs={"topic":"The future of electrical vehicles"})
```
